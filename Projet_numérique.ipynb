L'objecitf du projet est de déterminer le potentiel électrique $V$ autour d'une plaque $\Omega$ de dimensions finies, et nous allons y parvenir par analyse du problème.
Il s'agit du problème de Laplace, qu'on peut écire de la manière suivante :

$$\Delta V = 0, \text{ dans } \Omega$$
$$V \text{ constant sur la plaque}$$
$$\frac{\partial V}{\partial n} = 0, \text{ dans } \partial\Omega, \text{ où } n \text{ est une variable de } V$$

Mais il n'existe pas de solution analytique à ce problème. 
Pourtant, il est possible de trouver une solution qui s'en approche. Pour cela, nous considérerons qu'en un point proche d'un bord de la plaque, les effets des autres bord seront négligeables. Nous prendrons donc une demi-plaque infinie, qui dans un repère à deux dimensions aura l'allure d'une demi-droite.

### 1. Première solution

#### 1.2. Théorie

Pour répondre à ce nouveau problème, nous allons d'abord prendrons la fonction complexe $F$ définie de la manière suivant :
$$F(z) = z^\frac 12$$
Nous allons montrer que cette fonction permet de modéliser le potentiel autour de la plaque.
Comme $z$ et $F(z)$ sont des complexes, on peut écrire :
$$z=x+iy, \ \ \ \ F(x,y)=U(x,y)+iV(x,y)$$
où $U$ et $V$ sont deux fonctions réelles des variables $x$ et $y$.

Nous supposerons que le théorème mathématique suivant est vrai :
$$\frac {\partial U}{\partial x} = \frac {\partial V}{\partial y}$$
$$\frac {\partial U}{\partial y} = - \frac {\partial V}{\partial x}$$
On a alors :
$$\frac {\partial^2 U}{\partial x^2} = \frac {\partial}{\partial x} \left( \frac{\partial U}{\partial x} \right) = \frac {\partial}{\partial x} \left( \frac {\partial V}{\partial y} \right) = \frac {\partial^2 V}{\partial x \partial y}$$
$$\frac {\partial^2 U}{\partial y^2} = \frac {\partial}{\partial y} \left( \frac{\partial U}{\partial y} \right) = \frac {\partial}{\partial y} \left( -\frac {\partial V}{\partial x} \right) = -\frac {\partial^2 V}{\partial x \partial y}$$
$$\Longrightarrow \frac {\partial^2 U}{\partial x^2} + \frac {\partial^2 U}{\partial y^2} = \Delta U = 0$$
De même, 
$$\frac {\partial^2 V}{\partial x^2} = \frac {\partial}{\partial x} \left( \frac{\partial V}{\partial x} \right) = \frac {\partial}{\partial x} \left( -\frac {\partial U}{\partial y} \right) = -\frac {\partial^2 U}{\partial x \partial y}$$
$$\frac {\partial^2 V}{\partial y^2} = \frac {\partial}{\partial y} \left( \frac{\partial V}{\partial y} \right) = \frac {\partial}{\partial y} \left( \frac {\partial U}{\partial x} \right) = \frac {\partial^2 U}{\partial x \partial y}$$
$$\Longrightarrow \frac {\partial^2 V}{\partial x^2} + \frac {\partial^2 V}{\partial y^2} = \Delta V = 0$$

Les fonctions $U$ et $V$ remplissent donc la première condition au problème de Laplace. C'est d'ailleurs vrai pour toute fonction F complexe.
Pour les deux autres conditions, nous devrons exprimer explicitement $U$ et $V$.

On a :
$$F(z) = z^{\frac 12}$$
Comme $z= x + iy = \rho e^{i\theta}$ avec $\rho = (x^2 +y^2)^{\frac 12}$ et $\tan \theta = \frac yx,$
$$F(z) = (\rho e^{i\theta})^{\frac 12}$$
$$F(z) = \rho^{\frac 12} e^{i\frac \theta 2}$$
$$F(z) = \rho^{\frac 12} (\cos{\frac \theta 2} + i\sin{\frac \theta 2})$$
Or
$$\cos {\theta} = \cos(2 \cdot \frac{\theta}{2}) = 2\cos^2(\frac\theta 2) - 1 = \frac x\rho \Longrightarrow \cos{\frac \theta 2} = \left( \frac{x/\rho+1}{2} \right)^{\frac 12}$$
$$\cos {\theta} = 2\cos^2(\frac\theta 2) - 1 = 1 - 2\sin^2(\frac\theta 2)= \frac x\rho \Longrightarrow \sin{\frac \theta 2} = \left( \frac{1-x/\rho}{2} \right)^{\frac 12}$$
Donc
$$F(z) = \rho^{\frac 12} \left(\left( \frac{1 + x/\rho}{2} \right)^{\frac 12} + i\left( \frac{1-x/\rho}{2} \right)^{\frac 12}\right)$$
$$F(z) = \left( \frac{\rho+x}{2} \right)^{\frac 12} + i\left( \frac{\rho-x}{2} \right)^{\frac 12}$$
$$F(z) = \left( \frac{(x^2 +y^2)^{\frac 12}+x}{2} \right)^{\frac 12} + i\left( \frac{(x^2 +y^2)^{\frac 12}-x}{2} \right)^{\frac 12}$$
$$U(x,y) = \left( \frac{(x^2 +y^2)^{\frac 12}+x}{2} \right)^{\frac 12} \ ,\ \ \ V(x,y)=\left( \frac{(x^2 +y^2)^{\frac 12}-x}{2} \right)^{\frac 12}$$

Pour valider la deuxième consition, nous devons déterminer la position de la plaque dans l'espace. Nous avons dit précédemment qu'il s'agissait d'une demi-droite.
Dans le cas de $V$, c'est la demi-droite $y=0$ et $x≥0$. Dans le cas de $U$, c'est la demi-droite $y=0$ et $x≤0$.
Nous pouvons le vérifier de la manière suivante :
$$U(x,0) = \left( \frac{(x^2 +0^2)^{\frac 12}+x}{2} \right)^{\frac 12} = \left( \frac{|x|+x}{2} \right)^{\frac 12} = 0 \text{  si } x≤0$$
$$V(x,0) = \left( \frac{(x^2 +0^2)^{\frac 12}-x}{2} \right)^{\frac 12} = \left( \frac{|x|-x}{2} \right)^{\frac 12} = 0 \text{  si } x≥0$$

Pour la troisième condition, il faut calculer les dérivées partielles de $U$ et $V$ :
$$\frac{\partial U}{\partial x} = \frac{\partial V}{\partial y} = \frac{\sqrt{2}}{4(x^2+y^2)^{\frac 12}}\cdot \frac{y}{(x^2+y^2)^{\frac 12}-x} = 0 \text{   si  } y= 0$$
$$\frac{\partial V}{\partial x} = -\frac{\partial U}{\partial y} = \frac{\sqrt{2}}{4(x^2+y^2)^{\frac 12}}\cdot \frac{y}{(x^2+y^2)^{\frac 12}+x} = 0 \text{   si  } y= 0$$

Donc les fonctions $U$ et $V$ vérifient bien les conditions de Laplace. 
Nous choisirons la fonction $V$ comme potentiel électrique. 

#### 1.2. Équipotentielles

Maintenant, nous allons coder dans python les courbes équipotentielles. Ci-dessous les fonctions $U$ et $V$, ainsi que les valeurs de $y$ pour $x$ et $V$ fixés.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(x,y):                                                  #Fonction U de x et y, qui représente les lignes de champs
    U = np.sqrt((np.sqrt(x ** 2 + y ** 2) + x) / 2)
    return U

def V(x,y):                                                  #Renvoies le potentiel électrique V en fonction de x et y
    V = np.sqrt((np.sqrt(x ** 2 + y ** 2) - x) / 2)
    return V

def y(x,V):                                                  #Renvoie la position positive y en fonction de V et de x
    if V ** 2 + x <= 0:
        y = 0
    elif V ** 2 + x > 0:
        y = 2 * V * np.sqrt(V ** 2 + x)
    return y
```

```python
V = 2
L = 10
s = 0.01
```

```python
X = np.arange(-V**2, L, s)
Y0 = []
for i in X:
    Y0.append(y(i,V))
Y1 = []
for i in X:
    Y1.append(-y(i,V))

plt.plot([0,10],[0,0], color = 'black', label = 'plaque')
plt.plot(X, Y0, color = 'blue', label = 'équipotentiel')
plt.plot(X, Y1, color = 'blue')
plt.title('Courbe équipotentiel V = '+ str(V))
plt.legend()
plt.axis([-V**2,L,y(L,V),-y(L,V)])
plt.grid()
plt.show()
```

### 2. Deuxième solution

#### 2.1. Théorie

Nous allons maintenant considérer
$F(z) = z^{2/3}$
et montrer que cette fonction permet aussi de modéliser le potentiel autour de la plaque. 

Pour exprimer $U$ et $V$, on utilise le fait que :
$$z^{3/2} = z \cdot z^{1/2}$$

En posant $U_{1/2}$ et $V_{1/2}$ les fonctions obtenues pour la première solution ($z^{1/2}$), on a :
$$F(z) = (x + iy) \cdot (U_{1/2} + iV_{1/2})$$
$$F(z) = x U_{1/2} + i x V_{1/2} + i y U_{1/2} - y V_{1/2}$$
$$F(z) = (x U_{1/2} - y V_{1/2}) + i(x V_{1/2} + y U_{1/2})$$

Par identification, on obtient l'expression de $U(x,y)$ :
$$U(x,y) = x \left( \frac{(x^2 +y^2)^{1/2}+x}{2} \right)^{1/2} - y \left( \frac{(x^2 +y^2)^{1/2}-x}{2} \right)^{1/2}$$

et l'expression de $V(x,y)$ (Potentiel électrique) :
$$V(x,y) = x \left( \frac{(x^2 +y^2)^{1/2}-x}{2} \right)^{1/2} + y \left( \frac{(x^2 +y^2)^{1/2}+x}{2} \right)^{1/2}$$

Or $\rho = (x^2 + y^2)^{1/2}$ donc on peut simplifier l'écriture:

$$U(x,y) = x \left( \frac{\rho + x}{2} \right)^{1/2} - y \left( \frac{\rho - x}{2} \right)^{1/2}$$

$$V(x,y) = x \left( \frac{\rho - x}{2} \right)^{1/2} + y \left( \frac{\rho + x}{2} \right)^{1/2}$$

$$ \text{(noter que l'écriture $\rho$ nous sera utile pour la programmation)} $$

Il nous reste à verifier qu'elle vérifie les conditions du problème de Laplace

Les 2 premières conditions fonctionnent, d'ailleurs la preumière fonctionne pour toute fonction complexe U et V.

Ensuite la troisième condition:

On calcule $\frac{\partial U}{\partial x}$ :

$$\frac{\partial U}{\partial x} = \frac{\sqrt{2}}{4(x^2+y^2)^{\frac 12}}\cdot \frac{y}{(x^2+y^2)^{\frac 12}-x} x + \left( \frac{(x^2 +y^2)^{1/2}+x}{2} \right)^{1/2} + ...$$

condition = 0 si $y=0$

or si y = 0:

$$\frac{\partial U}{\partial x} = 0 \cdot x + \left( \frac{(x^2 +0)^{1/2}+x}{2} \right)^{1/2} + ...$$

On a donc besoin d'une condition supplémentaire sur x.


**Conclusion:** on ne peux pas prendre n'importe quelle fonction F(z) !

#### 2.2. Équipotentielles

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(x,y):                                                  #Fonction U de x et y, qui représente les lignes de champs
    U = np.sqrt((np.sqrt(x ** 2 + y ** 2) + x) / 2)
    return U

def V(x,y):                                                  #Renvoies le potentiel électrique V en fonction de x et y
    V = np.sqrt((np.sqrt(x ** 2 + y ** 2) - x) / 2)
    return V

def y(x,V):                                                  #Renvoie la position positive y en fonction de V et de x
    coeffs = [1, 3 * x, 0, -(4 * x**3 + 2 * V**2)]
    
    # On cherche les racines du polynôme
    racines = np.roots(coeffs)
    
    # On ne garde que la racine réelle qui est physiquement possible (rho >= |x|)
    real_roots = racines[np.isreal(racines)].real
    rho = real_roots[real_roots >= np.abs(x) - 1e-9]
    
    if len(rho) > 0:
        rho_val = rho[0]
        # On calcule y via Pythagore : y = sqrt(rho^2 - x^2)
        # Le max(0, ...) évite les erreurs de précision numérique (ex: -1e-15)
        return np.sqrt(max(0, rho_val**2 - x**2))
    else:
        return 0
```

```python

V = 2
L = 10
s = 0.01
```

```python
X = np.arange(-V**2, L, s)
Y0 = []
for i in X:
    Y0.append(y(i,V))
Y1 = []
for i in X:
    Y1.append(-y(i,V))

plt.plot([0,10],[0,0], color = 'black', label = 'plaque')
plt.plot(X, Y0, color = 'blue', label = 'équipotentiel')
plt.plot(X, Y1, color = 'blue')
plt.title('Courbe équipotentiel V = '+ str(V))
plt.legend()
plt.axis([-V**2,L,y(L,V),-y(L,V)])
plt.grid()
plt.show()
```
